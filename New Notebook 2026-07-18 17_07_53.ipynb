```python
data = [
    ("Ana", "Ventas", 3200),
    ("Luis", "Ventas", 2800),
    ("Marta", "Marketing", 3500),
    ("Carlos", "Marketing", 3100),
    ("Sofia", "TI", 4200),
]
columns = ["nombre", "area", "salario"]

df = spark.createDataFrame(data, columns)
df.show()

from pyspark.sql import functions as F

df_resumen = (
    df.groupBy("area")
      .agg(
          F.avg("salario").alias("salario_promedio"),
          F.count("*").alias("num_empleados")
      )
      .orderBy(F.desc("salario_promedio"))
)

df_resumen.show()

df.createOrReplaceTempView("empleados")

spark.sql("""
    SELECT area, ROUND(AVG(salario), 2) AS salario_promedio
    FROM empleados
    GROUP BY area
    ORDER BY salario_promedio DESC
""").show()
```
